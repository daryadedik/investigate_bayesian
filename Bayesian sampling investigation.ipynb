```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Data generation:
    we generate 1 million samples from Gaussian distribution

```python
num_points = 1000000
```

```python
def generate_data_normal(N):
    mu = np.random.uniform(-15, 15, 1) # we may use fixed mean and variance
    var = np.random.uniform(-1, 1, 1)
    xdata = 100 * np.random.random(N)
    ydata = mu + var * xdata  
    return (np.random.normal(xdata, 10), np.random.normal(ydata, 10))
```

```python
def plot_points(x, y):
    plt.figure(figsize=(10,5))
    plt.plot(x, y, 'ok', markersize=4)
    plt.grid()
    plt.show()
```

```python
np.random.seed(42)
xdata, ydata = generate_data_normal(num_points)
plot_points(xdata, ydata)
```
